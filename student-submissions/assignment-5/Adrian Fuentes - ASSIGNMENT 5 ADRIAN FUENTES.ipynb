i've notice that there is a lot of information with a lot of null values so i will impor the file again but this time only with the informaiton that i require and taking off the null values

```python
df= pd.read_csv('CLIWOC15.csv', usecols=['Lat3', 'Lon3', 'Nationality','Month','Year', "WarsAndFights", "WarsAndFightsMemo", "Cargo", "ShipName", "ShipType"])
df.dropna(axis=0, how='all', inplace=True)    ####i've choosen "all" because by chosing "any" i end with almost no data
df
```

```python
dfWAR = df[df.WarsAndFights==1]   ### now i only want the information than involves a fight
```

```python
dfWAR
```

to know wich army was more succesfull fighting i will count how many times did every army reported a conflict and because all this fights happened during a time where there was no telephone im assuming that the ships that reported the incident are the ones who won the battle.

```python
dfWAR.Nationality.value_counts()
```

Now i'll make a representation of where the battles happened

```python
dfbritish = dfWAR[dfWAR.Nationality=="British"]
dfdutch = dfWAR[dfWAR.Nationality=="Dutch"]
dffrench = dfWAR[dfWAR.Nationality=="French"]
dfdanish = dfWAR[dfWAR.Nationality=="Danish"]
dfspanish = dfWAR[dfWAR.Nationality=="Spanish"]
```

```python
fig = plt.figure(figsize=(20, 14))      ### this determines the proportion of the map
markersize = 100
markertype = 'D'
markeralpha = .4

m = Basemap(projection='merc',llcrnrlat=-80,urcrnrlat=80,\
        llcrnrlon=-180,urcrnrlon=180,lat_ts=20,resolution='c') 
m.drawcoastlines()

m.drawmapboundary(fill_color='white', linewidth=.01)
### british

markercolor = 'blue'

x, y = m(dfbritish.Lon3.values, dfbritish.Lat3.values)
m.scatter(x, y, markersize, marker=markertype, color=markercolor, alpha=markeralpha)

### dutch

markercolor = 'green'

x, y = m(dfdutch.Lon3.values, dfdutch.Lat3.values)
m.scatter(x, y, markersize, marker=markertype, color=markercolor, alpha=markeralpha)

#### french

markercolor = 'red'

x, y = m(dffrench.Lon3.values, dffrench.Lat3.values)
m.scatter(x, y, markersize, marker=markertype, color=markercolor, alpha=markeralpha)

###danish

markercolor = 'cyan'

x, y = m(dfdanish.Lon3.values, dfdanish.Lat3.values)
m.scatter(x, y, markersize, marker=markertype, color=markercolor, alpha=markeralpha)

###spanish

markercolor = 'magenta'

x, y = m(dfspanish.Lon3.values, dfspanish.Lat3.values)
m.scatter(x, y, markersize, marker=markertype, color=markercolor, alpha=markeralpha)
plt.title("Location of the Battles", fontsize=45)
plt.show()
```
