```python
import pandas as pd, requests, time
from geopy.geocoders import GoogleV3

google_maps_api_key='WRITE IN HERE YOU API'     #the API obtained by the the googe link above

import osmnx as ox, networkx as nx, pandas as pd
%matplotlib inline
ox.config(log_file=True, log_console=True, use_cache=True)
```

```python
locations = pd.DataFrame()
locations['address'] = ['Mappin St, Sheffield, UK S1 4DT',
                        'Bolsover, Sheffield, UK S3 7NA',
                        '32 Leavygreave RD, Sheffield, UK S3 7RD']
locations
```

```python
# function that accepts an address string, sends it to the Google API, and returns the lat-long API result
def geocode(address):
    url = 'http://maps.googleapis.com/maps/api/geocode/json?address={}&sensor=false' #api url with placeholders
    request = url.format(address) #fill in the placeholder with a variable
    response = requests.get(request) #send the request to the server and get the response
    data = response.json() #convert the response json string into a dict
    
    if len(data['results']) > 0: #if google was able to geolocate our address, extract lat-long from result
        latitude = data['results'][0]['geometry']['location']['lat']
        longitude = data['results'][0]['geometry']['location']['lng']
        return '{},{}'.format(latitude, longitude) #return lat-long as a string in the format google likes
```

```python
# for each value in the address column, geocode it, save results as new df column
locations['latlng'] = locations['address'].map(geocode)
locations
```

```python
# parse the result into separate lat and lon columns for easy mapping
locations['latitude'] = locations['latlng'].map(lambda x: x.split(',')[0])  ### in the list "location set the firs value of latlng as latitua
locations['longitude'] = locations['latlng'].map(lambda x: x.split(',')[1])  ### set the second value into longitude
locations
```

```python
#### Search for places near one the Arts Tower using a keyword

# google places api url, with placeholders
url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?keyword={}&location={}&radius={}&key={}&sensor=false'

# what keyword to search for
keyword = 'Restaurant'

# define the radius (in meters) for the search
radius = 300

location = locations.loc[1, 'latlng']
location
```

```python
# add our variables into the url, submit the request to the api, and load the response
request = url.format(keyword, location, radius, google_maps_api_key)
response = requests.get(request)
data = response.json()

### the .json response from google place api is limited to rturn up to 5 places
```

```python
### PLOT THE RESULTS FROM GOOGLE PLACES API (it is possible to retrieve more information, like the rating, vicinity, opening hours)

places = pd.DataFrame(data['results'])
places = places[['name', 'geometry', 'rating', 'vicinity', "opening_hours"]]  #https://developers.google.com/places/web-service/search
places.head()
```

```python
# parse out lat-long and return it as a series -> this creates a dataframe of all the results when you .apply()
def parse_coords(geometry):
    if isinstance(geometry, dict):
        lng = geometry['location']['lng']
        lat = geometry['location']['lat']
        return pd.Series({'latitude':lat, 'longitude':lng})
    
# test our function
places['geometry'].head().apply(parse_coords)
```

```python
# now run our function on the whole dataframe and SAVE THE OUTPUT to 2 new dataframe columns
places[['latitude', 'longitude']] = places['geometry'].apply(parse_coords)
places_clean = places.drop('geometry', axis=1)
places_clean.sort_values(by='rating', ascending=False).head()
```

```python
# define a point at the corner of California St and Mason St in SF
location_point = (53.3829744,-1.4870782)

# create network from point, inside bounding box of N, S, E, W each 750m from point
G2 = ox.graph_from_point(location_point, distance=500, distance_type='bbox', network_type='drive')
G2 = ox.project_graph(G2)
fig, ax = ox.plot_graph(G2, node_size=20, node_color='#66cc66')

```

```python
G = ox.graph_from_point(location_point, distance=600, distance_type='network', network_type='walk')
# define origin/desination points then get the nodes nearest to each
origin_point = (53.3829744,-1.4870782)
destination_point = (53.382111, -1.482613)
origin_node = ox.get_nearest_node(G, origin_point)
destination_node = ox.get_nearest_node(G, destination_point)

# find the route between these nodes then plot it
route = nx.shortest_path(G, origin_node, destination_node)
fig, ax = ox.plot_graph_route(G, route)
```
